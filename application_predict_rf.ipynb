# すべて同じ値の列は消去

```python
# 各行においてユニークな値のリスト
nunipue_count_list = list(a_dropna.nunique(dropna=False))
# 列名のリスト
a_dropna_columns_list = a_dropna.columns.values
# ユニークな値が1つだけの行を除外する
a_dropna_2 = a_dropna.copy()

for index in range(len(nunipue_count_list)):
#     ユニークな値が1つだけの時
    if nunipue_count_list[index] == 1:
        column_name = a_dropna_columns_list[index]
        a_dropna_2 = a_dropna_2.drop([column_name], axis=1)
        
a_dropna_2.nunique(dropna=False)
```

# NANが10000以上の指標を削除してみる

```python
# nanのカウント
null_count = list(a_dropna_2.isnull().sum())
# 列名のリスト
a_dropna_2_columns_list = a_dropna_2.columns.values
# NANが10000以上の指標を削除してみる
a_dropna_3 = a_dropna_2.copy()

for index in range(len(null_count)):
#     NANが10000以上のとき
    if null_count[index] >= 10000:
        column_name = a_dropna_2_columns_list[index]
        a_dropna_3 = a_dropna_3.drop([column_name], axis=1)
        
a_dropna_3.isnull().sum()
```

### メモ  
掲載期間開始日/掲載期間終了日：2種類しかないので0,1（普通のエンコーディング）

開始日と終了日は対応するので、終了日の方は削除へ

勤務地：拠点番号よりも詳しい（最悪省略でも）、東京・静岡などはグルーピングしたうえでエンコーディングできるのか

拠点番号：前の番号だけ抜粋してそのまま値に

勤務開始日：要検討→削除へ

勤務地　最寄り駅（沿線名）：勤務地と被ってくる→削除へ

勤務地最寄り駅：要検討→削除へ

# 回帰問題へ

データの分割

```python
y = df_y["応募数 合計"]
X = application_after.drop(["お仕事No."], axis=1)
```

```python
y_array = np.array(y)
X_array = np.array(X)
```

```python
print("y_array: ", y_array.shape)
print("X_array: ", X_array.shape)
```

```python
# train/testに分割する
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_array, y_array, test_size=0.4, random_state=0)
```

```python
print("X_train: ", X_train.shape)
print("y_train: ", y_train.shape)
print("X_test: ", X_test.shape)
print("y_test: ", y_test.shape)
```

```python
# ランダムフォレスト
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=0)
rfr.fit(X_train, y_train)
```

```python
y_pred = rfr.predict(X_test)
y_pred.shape
```

評価結果！

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_pred, y_test))
```
