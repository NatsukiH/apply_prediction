# 前処理

```python
import numpy as np
import pandas as pd
import sklearn
```

```python
# 読み込み
application = pd.read_csv("train_x.csv")
application.head()
```

```python
application.describe()
```

### 全てnanのcolumnを削除してみる

```python
# 全てnanのcolumnを削除
a_1 = application.dropna(how='all', axis=1)
a_1.head()
```

```python
a_1.nunique(dropna=False)
```

### すべて同じ値の列は消去してみる

```python
# 各行においてユニークな値のリスト
nunipue_count_list = list(a_1.nunique(dropna=False))
# 列名のリスト
a_1_columns_list = a_1.columns.values
# ユニークな値が1つだけの行を除外する
a_2 = a_1.copy()

for index in range(len(nunipue_count_list)):
#     ユニークな値が1つだけの時
    if nunipue_count_list[index] == 1:
        column_name = a_1_columns_list[index]
        a_2 = a_2.drop([column_name], axis=1)
        
a_2.nunique(dropna=False)
```

```python
a_2.head()
```

### NANが10000以上の指標を削除してみる

```python
# nanのカウント
null_count = list(a_2.isnull().sum())
# 列名のリスト
a_2_columns_list = a_2.columns.values
# NANが10000以上の指標を削除してみる
a_3 = a_2.copy()

for index in range(len(null_count)):
#     NANが10000以上のとき
    if null_count[index] >= 10000:
        column_name = a_2_columns_list[index]
        a_3 = a_3.drop([column_name], axis=1)
        
a_3.isnull().sum()
```

```python
a_3.head()
```

### 長い文章のカラムを削除してみる

```python
a_4 = a_3.drop(["休日休暇　備考", "期間・時間　勤務時間", "お仕事名", "（派遣先）配属先部署", "仕事内容",  "応募資格", "派遣会社のうれしい特典", "お仕事のポイント（仕事PR）", "（派遣先）職場の雰囲気", "給与/交通費　備考"], axis=1)
a_4.head()
```

```python
a_4 = a_4.drop(["勤務地　最寄駅2（駅名）", "勤務地　最寄駅2（沿線名）"], axis=1)
a_4.head()
```

```python
# object型のみ抽出
a_4_object = a_4.select_dtypes(include=object)
a_4_object.head()
```

### メモ  
掲載期間開始日/掲載期間終了日：2種類しかないので0,1（普通のエンコーディング）

開始日と終了日は対応するので、終了日の方は削除へ

勤務地：拠点番号よりも詳しい（最悪省略でも）、東京・静岡などはグルーピングしたうえでエンコーディングできるのか

拠点番号：前の番号だけ抜粋してそのまま値に

勤務開始日：要検討→削除へ

勤務地　最寄り駅（沿線名）：勤務地と被ってくる→削除へ

勤務地最寄り駅：要検討→削除へ

```python
a_5 = a_4.drop(["勤務地　備考", "期間・時間　勤務開始日", "勤務地　最寄駅1（沿線名）", "勤務地　最寄駅1（駅名）", "掲載期間　終了日"], axis=1)
a_5.head()
```

```python
# 掲載開始日を０，１にエンコード
day_mapping = {"2019/11/27": 0, "2019/9/25": 1}
a_5["掲載期間　開始日"] = a_5['掲載期間　開始日'].map(day_mapping)
```

```python
func = lambda x: int(x[:3])
a_5["拠点番号"] = a_5["拠点番号"].map(func)
a_5.head()
```

```python
a_5.describe()
```

### NANの消去

```python
a_6 = a_5.fillna(-99)
a_6.isnull().sum()
```

```python
application_after = a_6
application_after.head()
```

### yの読み込み

```python
df_y = pd.read_csv("train_y.csv")
df_y.head()
```

```python
len(df_y)
```

# 回帰問題

```python
y = df_y["応募数 合計"]
X = application_after.drop(["お仕事No."], axis=1)
```

```python
y_array = np.array(y)
X_array = np.array(X)
```

```python
print("y_array: ", y_array.shape)
print("X_array: ", X_array.shape)
```

```python
# train/testに分割する
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_array, y_array, test_size=0.4, random_state=0)
```

```python
print("X_train: ", X_train.shape)
print("y_train: ", y_train.shape)
print("X_test: ", X_test.shape)
print("y_test: ", y_test.shape)
```

### ランダムフォレスト

```python
# ランダムフォレスト
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=0)
rfr.fit(X_train, y_train)
```

```python
y_pred = rfr.predict(X_test)
y_pred.shape
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_pred, y_test))
```
