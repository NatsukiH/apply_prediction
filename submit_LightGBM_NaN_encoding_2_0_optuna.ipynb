### 全てnanのcolumnを削除してみる

```python
# 全てnanのcolumnを削除
a_1 = application.dropna(how='all', axis=1)
a_1.head()
```

```python
a_1.nunique(dropna=False)
```

### すべて同じ値の列は消去してみる

```python
# 各行においてユニークな値のリスト
nunipue_count_list = list(a_1.nunique(dropna=False))
# 列名のリスト
a_1_columns_list = a_1.columns.values
# ユニークな値が1つだけの行を除外する
a_2 = a_1.copy()

for index in range(len(nunipue_count_list)):
#     ユニークな値が1つだけの時
    if nunipue_count_list[index] == 1:
        column_name = a_1_columns_list[index]
        a_2 = a_2.drop([column_name], axis=1)
        
a_2.nunique(dropna=False)
```

```python
a_2.head()
```

### NANが10000以上の指標を削除してみる

```python
# nanのカウント
null_count = list(a_2.isnull().sum())
# 列名のリスト
a_2_columns_list = a_2.columns.values
# NANが10000以上の指標を削除してみる
a_3 = a_2.copy()

for index in range(len(null_count)):
#     NANが10000以上のとき
    if null_count[index] >= 10000:
        column_name = a_2_columns_list[index]
        a_3 = a_3.drop([column_name], axis=1)
        
a_3.isnull().sum()
```

```python
import optuna.integration.lightgbm as lgb_o
```

```python
# 特徴量と目的変数をlightgbmのデータ構造に変換する
lgb_train = lgb_o.Dataset(X_train, y_train)
lgb_eval = lgb_o.Dataset(X_test, y_test)
```

```python
# ハイパーパラメータの設定
# "colsample_bytree"を変更
params = {"colsample_bytree": 0.7, "seed": 71, "verbose": 0, 'objective': 'regression', 'metric': 'rmse'}
# num_round = 265
num_round = 100
```

#### パラメータチューニングに関するメモ  
max_depth: 増やしても精度が上がらなかった  
colsample_bytree: 小さい方が良いみたい(0.7より小さいと精度が上がらない)  

```python
# 学習の実行
# バリデーションデータもモデルに渡し、学習の進行とともにスコアがどう変わるかモニタリングする
model = lgb_o.train(params, lgb_train, num_boost_round = num_round, valid_names = ["train", "valid"], valid_sets = [lgb_train, lgb_eval])
```

```python
# バリデーションデータでのスコアの確認
va_pred = model.predict(X_test)
# 負の値は0へ
va_pred_0 = [0 if i <= 0 else i for i in va_pred]
```

```python
from sklearn.metrics import mean_squared_error
score = np.sqrt(mean_squared_error(va_pred_0, y_test))
print(score)
```

```python
# 予測
pred = model.predict(submit_x_test)
pred_0 = [0 if i <= 0 else i for i in pred]
```
