```python
q_x = pd.read_csv("test_x.csv")
q_x.head()
```

```python
# 全てnanのcolumnを削除
q_x_1 = q_x.dropna(how='all', axis=1)
q_x_1.head()
```

```python
# 各行においてユニークな値のリスト
nunipue_count_list = list(q_x_1.nunique(dropna=False))
# 列名のリスト
q_x_1_columns_list = q_x_1.columns.values
# ユニークな値が1つだけの行を除外する
q_x_2 = q_x_1.copy()

for index in range(len(nunipue_count_list)):
#     ユニークな値が1つだけの時
    if nunipue_count_list[index] == 1:
        column_name = q_x_1_columns_list[index]
        q_x_2 = q_x_2.drop([column_name], axis=1)
        
q_x_2.nunique(dropna=False)
```
